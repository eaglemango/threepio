```python
import requests
```

```python
response = None

with open("sentences.txt", "rt") as all_sentences:
	sentences = all_sentences.readlines()
	response = requests.post("http://0.0.0.0:3004/model", json={"sentences": sentences})

	print(response.json())
```

```python
len(response.json())
```

```python
json = response.json()
```

```python
json[0]
```

```python
json[0][0]
```

```python
anger = []
fear = []
joy = []
love = []
sadness = []
surprise = []
neutral = []
```

```python
for i in range(len(json)):
    anger.append(json[i][0]["anger"])
    fear.append(json[i][0]["fear"])
    joy.append(json[i][0]["joy"])
    love.append(json[i][0]["love"])
    sadness.append(json[i][0]["sadness"])
    surprise.append(json[i][0]["surprise"])
    neutral.append(json[i][0]["neutral"])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))

plt.hist(anger, bins=20, alpha=0.5)
plt.hist(fear, bins=20, alpha=0.5)
plt.hist(joy, bins=20, alpha=0.5)
plt.hist(love, bins=20, alpha=0.5)
plt.hist(sadness, bins=20, alpha=0.5)
plt.hist(surprise, bins=20, alpha=0.5)
# plt.hist(neutral, alpha=0.2)

plt.legend(["anger", "fear", "joy", "love", "sadness", "surprise"])

plt.show()
```
